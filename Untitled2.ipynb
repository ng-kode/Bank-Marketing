```python
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import math


font = {'family' : 'normal',
        'weight' : 'bold'}
matplotlib.rc('font', **font)

%matplotlib inline
```

```python
# Source of dataset: http://archive.ics.uci.edu/ml/datasets/Bank+Marketing

df = pd.read_csv('bank-additional-full.csv', sep=';')
df.head()
```

```python
df.columns
```

```python
# duration should not be included as that was unknown before contact complete
df_features = df[['age', 'job', 'marital', 'education', 'default', 'housing', 'loan',
       'contact', 'month', 'day_of_week','campaign', 'pdays',
       'previous', 'poutcome', 'emp.var.rate', 'cons.price.idx',
       'cons.conf.idx', 'euribor3m', 'nr.employed']]
```

```python
df_labels = df[['y']]
```

```python
sns.countplot(df['y'])
```

```python
plt.figure(figsize=(20, 50))

nrows = math.ceil(len(df_features.columns) / 3)
ncols = 3
categorical_features = []

for i, col in enumerate(df_features.columns):
    try:
        position = i + 1
        plt.subplot(nrows, ncols, position)
        sns.distplot(df[col])
    except TypeError:
        categorical_features.append(col)
        sns.countplot(x=col, hue='y', data=df)
    finally:
        plt.xticks(rotation=45)
```

### Data preprocessing

```python
categorical_features
```

```python
df_features = pd.concat([df_features.drop(categorical_features, axis=1), pd.get_dummies(df_features[categorical_features], drop_first=True)], axis=1)
```

```python
df_labels = pd.get_dummies(df_labels)[['y_yes']].rename(columns={'y_yes':'y'})
```

### Train-test split

```python
X_train, X_test, y_train, y_test = train_test_split(df_features, df_labels, test_size=0.33, random_state=0)
```

### Normalize data

```python
# numerical_features = [feat for feat in df.columns if feat not in categorical_features and feat != 'y' and feat != 'duration']

# X_mean = X_train[numerical_features].mean(axis=0)
# X_std = X_train[numerical_features].std(axis=0)

X_mean = X_train.mean(axis=0)
X_std = X_train.std(axis=0)
```

```python
# X_train[numerical_features] = (X_train[numerical_features] - X_mean) / X_std
# X_test[numerical_features] = (X_test[numerical_features] - X_mean) / X_std

X_train = (X_train - X_mean) / X_std
X_test = (X_test - X_mean) / X_std
```

### Train-validation split

```python
validation_samples = int(len(X_train) * 0.33)

X_val = X_train[:validation_samples]
y_val = y_train[:validation_samples]
partial_X_train = X_train[validation_samples:]
partial_y_train = y_train[validation_samples:]
```

### Overview of some models

```python
from sklearn.metrics import confusion_matrix

def train_and_eval(classifier, title):
    print('Start training %s' % title)
    classifier.fit(partial_X_train, partial_y_train.values.flatten())    
    y_pred_val = classifier.predict(X_val)
    
    cm = confusion_matrix(y_val, y_pred_val)
    
    print('Confusion Matrix, %s' %  title)
    print(cm)
    print('\n')
    return cm

results = dict()

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
results['DecisionTreeClassifier'] = train_and_eval(dtc, 'Decision Tree Classifier')

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
results['naive_bayes'] = train_and_eval(nb, 'Naive Bayes Classifier')

from sklearn.svm import SVC
svm = SVC()
results['kernel svm'] = train_and_eval(svm, 'Kernel SVM')

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
results['RandomForestClassifier'] = train_and_eval(rfc, 'RandomForestClassifier')

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
results['KNeighborsClassifier'] = train_and_eval(knn, 'KNeighborsClassifier')
```

```python
import itertools

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        # print("Normalized confusion matrix")
    # else:
        # print('Confusion matrix, without normalization')

    # print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
from math import ceil

ncols = 3
nrows = ceil(len(results) / ncols)

plt.figure(figsize=(20, 12))
for i, key in enumerate(results):
    plt.subplot(nrows, ncols, i + 1)
    plot_confusion_matrix(cm=results[key], classes=y_val['y'].unique(), normalize=True, title='Normalized Confusion matrix, %s' % key)
```

```python
from sklearn.metrics import precision_recall_curve, auc
import matplotlib.pyplot as plt
from math import ceil

plt.figure(figsize=(12, 6))
colors = ['b', 'g', 'r', 'c', 'm']
labels = ['dtc', 'nb', 'svm', 'rfc', 'knn']
for i, clf in enumerate([dtc, nb, svm, rfc, knn]):    
    if clf is svm:
        y_score = clf.decision_function(X_val)
    else:
        y_score = clf.predict_proba(X_val)[:, 1]
        
    precision, recall, _ = precision_recall_curve(y_val, y_score)

    plt.step(recall, precision, color=colors[i], alpha=0.2,
             where='post', label=labels[i])
    plt.fill_between(recall, precision, step='post', alpha=0.2,
                     color=colors[i])
    
    print('Area under curve, {}: {}'.format(labels[i], auc(recall, precision)))
    
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('2-class Precision-Recall curves') 
plt.legend()
```

```python
### Upsampling
```

```python
# duration should not be included as that was unknown before contact complete
categorical_features = ['age', 'job', 'marital', 'education', 'default', 'housing', 'loan',
       'contact', 'month', 'day_of_week','campaign', 'pdays',
       'previous', 'poutcome', 'emp.var.rate', 'cons.price.idx',
       'cons.conf.idx', 'euribor3m', 'nr.employed']
df_features = df[categorical_features]
df_labels = df[['y']]

df_features = pd.concat([df_features.drop(categorical_features, axis=1), pd.get_dummies(df_features[categorical_features], drop_first=True)], axis=1)
df_labels = pd.get_dummies(df_labels)[['y_yes']].rename(columns={'y_yes':'y'})

X_train, X_test, y_train, y_test = train_test_split(df_features, df_labels, test_size=0.33, random_state=0)

X_mean = X_train.mean(axis=0)
X_std = X_train.std(axis=0)
X_train = (X_train - X_mean) / X_std
X_test = (X_test - X_mean) / X_std

validation_samples = int(len(X_train) * 0.33)

X_val = X_train[:validation_samples]
y_val = y_train[:validation_samples]
X_train_res = X_train[validation_samples:]
y_train_res = y_train[validation_samples:]

print('Before upsampling')
print(y_train_res.y.value_counts())

from imblearn.over_sampling import SMOTE
sm = SMOTE(ratio = 1.0)
X_train_res, y_train_res = sm.fit_sample(X_train_res, y_train_res)

print('After upsampling')
print(np.unique(y_train_res, return_counts=True))

from sklearn.metrics import confusion_matrix

def train_and_eval(classifier, title):
    print('Start training %s' % title)
    classifier.fit(X_train_res, y_train_res)    
    y_pred_val = classifier.predict(X_val)
    
    cm = confusion_matrix(y_val, y_pred_val)
    
    print('Confusion Matrix, %s' %  title)
    print(cm)
    print('\n')
    return cm

results = dict()

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
results['DecisionTreeClassifier'] = train_and_eval(dtc, 'Decision Tree Classifier')

from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
results['naive_bayes'] = train_and_eval(nb, 'Naive Bayes Classifier')

from sklearn.svm import SVC
svm = SVC()
results['kernel svm'] = train_and_eval(svm, 'Kernel SVM')

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
results['RandomForestClassifier'] = train_and_eval(rfc, 'RandomForestClassifier')

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
results['KNeighborsClassifier'] = train_and_eval(knn, 'KNeighborsClassifier')
```

```python
from math import ceil

ncols = 3
nrows = ceil(len(results) / ncols)

plt.figure(figsize=(20, 12))
for i, key in enumerate(results):
    plt.subplot(nrows, ncols, i + 1)
    plot_confusion_matrix(cm=results[key], classes=y_val['y'].unique(), normalize=True, title='Normalized Confusion matrix, %s' % key)
```
