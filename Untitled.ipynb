```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
import math
```

```python
# Source of dataset: http://archive.ics.uci.edu/ml/datasets/Bank+Marketing

df = pd.read_csv('bank-additional-full.csv', sep=';')
df.head()
```

```python
# duration should not be included as that was unknown before contact complete
df_features = df[['age', 'job', 'marital', 'education', 'default', 'housing', 'loan',
       'contact', 'month', 'day_of_week','campaign', 'pdays',
       'previous', 'poutcome', 'emp.var.rate', 'cons.price.idx',
       'cons.conf.idx', 'euribor3m', 'nr.employed']]
```

```python
df_labels = df[['y']]
```

```python
sns.countplot(df['y'])
```

```python
plt.figure(figsize=(20, 50))

nrows = math.ceil(len(df_features.columns) / 3)
ncols = 3
categorical_features = []

for i, col in enumerate(df_features.columns):
    try:
        position = i + 1
        plt.subplot(nrows, ncols, position)
        sns.distplot(df[col])
    except TypeError:
        categorical_features.append(col)
        sns.countplot(x=col, hue='y', data=df)
```

```python
def pdays_value_convert(x):
    if x == 999:
        return -1
    else:
        return x

df_features['pdays'] = df_features['pdays'].apply(pdays_value_convert)
```

### Data preprocessing

```python
categorical_features
```

```python
df_features = pd.concat([df_features.drop(categorical_features, axis=1), pd.get_dummies(df_features[categorical_features], drop_first=True)], axis=1)
```

```python
len(df_features[df_features['pdays'] == -1]) / len(df_features)
```

```python
df_labels = pd.get_dummies(df_labels)[['y_yes']].rename(columns={'y_yes':'y'})
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_features, df_labels, test_size=0.33, random_state=0)
```

```python
X_train[numerical_features] = (X_train[numerical_features] - X_mean) / X_std
```
